```python
import pandas as pd
import numpy as np


data = pd.read_csv("raw_delivery_logs_50k.csv")
```

```python
abc = data.copy()
```

```python
abc.isna().sum()
```

```python
abc["order_timestamp"] = pd.to_datetime(abc["order_timestamp"])
abc["delivery_completion_timestamp"] = pd.to_datetime(abc["delivery_completion_timestamp"])

abc["delivery_time_mins"] = (abc["delivery_completion_timestamp"] - abc["order_timestamp"]).dt.total_seconds()/60

abc["delivery_time_mins"] = abc["delivery_time_mins"].round(1)

print("Missing values per column:")
print(abc.isna().sum())
```

```python

conditions = [
    abc['delivery_time_mins'].isna(),       
    abc['delivery_time_mins'] <= 15.0,      
    abc['delivery_time_mins'] > 15.0        
]

choices = ['System Error', 'On Time', 'Delayed']

abc['delivery_status'] = np.select(conditions, choices, default='Unknown')

print("Delivery Status Breakdown:")
print(abc['delivery_status'].value_counts())
```

```python

abc_clean = abc.drop(columns=['delivery_completion_timestamp'])


abc_clean.to_csv('cleaned_deliveries_50k.csv', index=False)

print("Data Cleaning Complete. Ready for SQL Database.")
```
